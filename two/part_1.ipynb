##### Homography is a projective transformation that relates two images of same planar surface under different perspectives. By defining at least four point correspondence between the distorted image and a reference rectangular image, the homography matrix H (a 3x3 transformation matrix) maps each pixel in the Source image to its geometrically correct position.
##### Applying the inverse homography warps the distorted image into the rectified coordinate system, this confirms the goal of rectification, which is to remove the effects of perspective distortion so that objects that are planar in the scene appear fronto-parallel in the image.
