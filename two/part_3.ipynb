### Triangular Mesh Warp
### Advantage: It is faster to compute
### Disadvantages: The quality of the warp depends on the distribution of the triangles

### Thin Plate Spline Warp

### Advantage: No triangulation required, unlike the triangular mesh warp

### Disavantage: Slower than Mesh Warp
